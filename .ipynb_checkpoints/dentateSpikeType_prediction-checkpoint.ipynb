# Tutorial on how to use dSpikesClassifier

dSpikesClassifier can predict the type of dentate spike (i.e. either DS1 or DS2)

To use it, you only need the local field potential (LFP) signal from the granule-cell layer of the dentate gyrus (DG) and the times of the dentate spikes.

In this tutorial, you will learn:
1. How to extract dentate spikes times from your LFP. For this, you need the LFP from an additional reference channel (here we use one in stratum oriens)

2. How to use dSpikesClassifier to pre-process the data and classify events into either DS1 or DS2.

At the end of the tutorial, we show how the model predictions match the currents profiles described for DS1 and DS2 (Bragin et al., 1995; Meier et al., 2020; Dvorak et al., 2021).

```python
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['xtick.labelsize']=15
matplotlib.rcParams['ytick.labelsize']=15

# Suppress the specific warning due to sklearn (model was saved in previous sklean version)
import warnings
warnings.filterwarnings("ignore", category=UserWarning, module="sklearn")

## import custom libraries
import utils
import dSpikes_model as dSpikesClassifier
```

## Load Data for git 

```python
resDir = './Data/'

# Load LFP of reference channel and LFP of DG channel
## Here we are using a channel in the stratum oriens as reference
lfp_ref = np.load(resDir+'mvl33-221109_3.lfp.ref',allow_pickle=True,mmap_mode='r')
lfp_dg = np.load(resDir+'mvl33-221109_3.lfp.dg',allow_pickle=True,mmap_mode='r')

sr = utils.sr # sampling rate

# Concatenate LFPs from different channels (this can be all your channels)
# Here we are using two ust for data storage constrains in the repo
lfps = np.concatenate((lfp_ref[None,:],lfp_dg[None,:]),axis=0)
lfps.shape 
```

```python
# Load Layers of Channels 
trodes = np.load(resDir+'mvl33-221109.channels',allow_pickle=True)
trodes
```

### Show example of lfp window

```python
# Plot DG LFP window
t = np.arange(0,sr) # 1 sec of data
t_sec = t/sr

lfp_win = lfps[:,t]
plt.figure(figsize=(12,3))
plt.plot(t_sec,lfp_win[0],label='Oriens',color='orange',lw=1.5)
plt.plot(t_sec,lfp_win[1],label='GCL',color='k',lw=1.5)
plt.xlabel('Time (s)',fontdict=utils.fontdict)
plt.ylabel('LFP (mV)',fontdict=utils.fontdict)
plt.legend()
utils.AdjustBoxPlot()
plt.show()
```

## Run DSpikes Prediction or Load them 

```python
dg_chi = 1# channel for dentate spikes predicion
# reference channel to compare dg lfp will be 0 by default
dspikes = dSpikesClassifier.runDSpikes_detection(lfps,dg_chi=dg_chi)

sdur_s = lfps.shape[1]/sr # session duration in seconds
print(f'\n\nDetected {len(dspikes)} dentate spikes.')
print('  Avg. rate of dentate spikes {} events/sec'.format(np.round(len(dspikes)/sdur_s,2)))
```

```python
## Trigger LFP in all channels with dentate spikes times 
win_trigger_ms = 200 # window size in ms used to trigger lfp for each event
win_trigger_samples = int((win_trigger_ms/1000)*sr) # convert to samples

lfp_dspikes,taxis = utils.triggeredAverage(lfps,dspikes,taLen=win_trigger_samples,average=False)

print(lfp_dspikes.shape)
print('(# dspikes, # channels, # time samples)')
```

```python
# plot DG LFP for example dentate spike
examplei = 0 #select event to plot

plt.figure(figsize=(5,3))
plt.plot(taxis,lfp_dspikes[examplei][dg_chi],color='k',lw=2)

plt.xlabel('Time from dentate spike (ms)',fontdict=utils.fontdict)
plt.ylabel('LFP (mV)',fontdict=utils.fontdict)
utils.AdjustBoxPlot()
plt.show()
```

## Run Model for DS1 and DS2 classification

```python
dSpikesClassifier.model_report()
```

### Classification by giving the LFP of all channels and the times of dspikes

```python
## The model will 
# 1) trigger the lfps in the GR with the dspikes timestamps
# 2) Choose the channel in dg which best matches the DS template used to train of model
# 3) Pre-Process LFP in selected channel for DS class prediction
# 4) Predict the Class of DS spikes (i.e. DS1 or DS2 for each event)

###### OUTPUTS #########################################################
#     DSclass_pred: DSspike class prediction  0s and 1s -> ds1,ds2
#     lfp_ds_GR: is the lfp in the DG reference channel triggered for all DSpikes
#     dspikes: DSpikes timestamps corresponding to the class 
#              (in case triggering window exceds lfp duration)

DSclass_pred,lfp_ds_GR,dspikes_c = dSpikesClassifier.predictDS_class(lfps,dspikes=dspikes\
                                                 ,GRchi=None,trodes=trodes,plot=True,report=True)
```

```python
## Or you can manually hoose the channel used for DS class prediction 
# By setting GRchi to a integer;this allows you to put trodes=None
chi = dg_chi # select DG channel for model 

DSclass_pred,lfp_ds_GR,dspikes_c = dSpikesClassifier.predictDS_class(lfps,dspikes=dspikes,\
                                                 GRchi=chi,trodes=None,plot=True,report=True)
```

```python
## You can also trigger the LFP before and give the dspikes-triggered LFP to the model

# Trigger LFPs in all channels with DSPikes times
lfp_ds = dSpikesClassifier.triggerLFP(lfps,dspikes)
print(lfp_ds.shape,' - (# spikes, # channels, # num samples (eeg))')
chi = 1 # select reference DG channel for classification

DSclass_pred,lfp_ds_GR,dspikes_c = dSpikesClassifier.predictDS_class(lfp_ds,GRchi=chi,
                                                       trodes=trodes,plot=True,report=True)
```

### Visualise Predictions in Lower Dimensional Space

```python
from sklearn.manifold import Isomap
# Perform Dimensionality reduction on DG LFP signatures during DS
embed = Isomap(n_components=2, metric='correlation',n_neighbors=40).fit(lfp_ds_GR)
proj = embed.transform(lfp_ds_GR)

# Plot events in lower-dimensional space, color-coding for the DS type label
# predicted by the model
plt.figure(figsize=(5,3))
for classi,color in enumerate(dSpikesClassifier.colors_DSs):
    mask = DSclass_pred == classi
    
    plt.scatter(proj[mask][:,0],proj[:,1][mask],color=color,\
                alpha=.75,label='DS'+str(classi+1))

plt.xlabel('Component 1 (au)',fontdict=utils.fontdict)
plt.ylabel('Component 2 (au)',fontdict=utils.fontdict)
plt.legend()
utils.AdjustBoxPlot()
plt.show()
```

### Validate Classes Prediction with CSD Profiles along CA1-DG Axis

Now, we want to show that the classification purely based on the LFP signatures withing the DG correlates with the current source density (CSD) profiles described for DS1 and DS2 (Bragin et al., 1995; Meier et al., 2020; Dvorak et al., 2021).

Due to storage capacity of the repository, we could not load the lfps from a 64-channels probe that we used to validate the model. Threfore, we pre-computed and stores the triggered average LFP for DS1 nd DS2 predicted by the dSpikesClassifier for all channels.

```python
# Loading the triggered average LFPs of DS1 and DS2 
# from all 64 channels triggered. DS1 and DS2 were predicted as above,
# by applying the dSpikesClassifier to the LFP from the DG channel 
trigLFP_DSs = np.load(resDir+'mvl33-221109_3.lfp.DS.avg',allow_pickle=True)

winL_ms = 200
talen = int(2 * (winL_ms / utils.DTT))
talenh = int(talen / 2)
taxis_ = utils.DTT * (np.arange(-talenh, talenh, 1))

tmask = np.digitize(taxis_,[-100,100])==1
taxis = taxis_[tmask]
# Loading layers corresponding to each channel
layers = np.load(resDir+'mvl33-221109.probe.layers',allow_pickle=True)

trigLFP_DSs.keys()
```

We can estimate the CSD  as (Mitzdorf, 1985):

$$CSD[t]_n=−(LFP[t]_{n−1}−2∗LFP[t]_n+LFP [t]_{n+1})$$

where $LFP[t]_{n−1}$, $LFP[t]_n$ and $LFP[t]_{n+1}$ are the LFP signals at time t recorded from neighboring electrodes (20 μm apart).

```python
# Set color levels for CSD plot
levels_ = np.linspace(-.1,.1,50)

plt.figure(figsize=(10,8))
# Plot CSD of Triggered Average LFP 
for i,lbl in enumerate(['DS1','DS2']):
    # Get triggered average LFP of DS1 or DS2 events
    trigAvg_lfp_ds =  trigLFP_DSs[lbl][:,tmask]
    
    # Compute CSD 
    csd_ds = utils.runCSD(trigAvg_lfp_ds,spacing=20)
    
    # Plot CSD
    plt.subplot(1,2,i+1)
    utils.plotCSD(csd_ds,trigAvg_lfp_ds,taxis,spacing=1,
                levels_=levels_,
                xlim=None,
                cmap='seismic',
                ax_lbl={'ylabel':'Layer','xlabel':'Time from '+lbl+' (ms)'},
                ax=plt.gca())
    utils.probe_yticks(layers)

plt.suptitle('Average CSD Profile of Model Estimated DS1 and DS2',fontsize=16)    
plt.tight_layout()
```
